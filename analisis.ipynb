```python
import pandas as pd
import matplotlib.pyplot as plt

# create a sample DataFrame
data = 'cleaned_data.csv'
df = pd.read_csv(data)

# create a dictionary to store the count of occurrences of 1 in each column
count_dict = {}

# iterate over each column in the list of columns
for col in ['HS', 'Abusive','HS_Individual','HS_Group','HS_Religion','HS_Race','HS_Physical','HS_Gender', 'HS_Other','HS_Weak','HS_Moderate','HS_Strong']:
    # count the number of occurrences of 1 in the current column
    count = df[col].value_counts()[1]

    # store the count in the count_dict dictionary
    count_dict[col] = count

# create a list of labels for the x-axis
labels = list(count_dict.keys())

# create a list of values for the y-axis
values = list(count_dict.values())

# create a bar chart
plt.bar(labels, values)

# add a title and labels to the axes
plt.title('Number of Occurrences of 1 in Each Column')
plt.xlabel('Column')
plt.ylabel('Number of Occurrences')

# rotate the x-axis labels
plt.xticks(rotation=45)

# add data labels to the bars
for i, v in enumerate(values):
    plt.text(i, v+3, str(v), ha='center')

# display the chart
plt.show()
```

```python
# filter the data to get rows where HS is 1 or 0
hs_1 = df[df['HS'] == 1]
hs_0 = df[df['HS'] == 0]

# calculate the percentage of rows with HS as 1 and 0
percentage_hs_1 = (hs_1.shape[0] / df.shape[0]) * 100
percentage_hs_0 = (hs_0.shape[0] / df.shape[0]) * 100

# create a dictionary to store the percentage of occurrences of 1 in each column
percentage_dict_hs = {'HS': percentage_hs_1, 'Non HS': percentage_hs_0}

# create a list of labels for the x-axis
labels_hs = list(percentage_dict_hs.keys())

# create a list of values for the y-axis
values_hs = list(percentage_dict_hs.values())

# create a pie chart for HS
plt.subplot(1, 2, 1)
plt.pie(values_hs, labels=labels_hs, autopct='%1.1f%%')
plt.title('Ujaran Kebencian (HS)')

# filter the data to get rows where Abusive is 1 or 0
abusive_1 = df[df['Abusive'] == 1]
abusive_0 = df[df['Abusive'] == 0]

# calculate the percentage of rows with Abusive as 1 and 0
percentage_abusive_1 = (abusive_1.shape[0] / df.shape[0]) * 100
percentage_abusive_0 = (abusive_0.shape[0] / df.shape[0]) * 100

# create a dictionary to store the percentage of occurrences of 1 in each column
percentage_dict_abusive = {'Abusive': percentage_abusive_1, 'Non Abusive': percentage_abusive_0}

# create a list of labels for the x-axis
labels_abusive = list(percentage_dict_abusive.keys())

# create a list of values for the y-axis
values_abusive = list(percentage_dict_abusive.values())

# create a pie chart for Abusive
plt.subplot(1, 2, 2)
plt.pie(values_abusive, labels=labels_abusive, autopct='%1.1f%%')
plt.title('Kata Kasar (Abusive)')

# display the charts
plt.show()
```

```python
# import necessary libraries

from wordcloud import WordCloud

# select rows where any of the columns ['HS', 'Abusive','HS_Individual','HS_Group'] are 1
selected_rows = df[(df['HS'] == 1) | (df['Abusive'] == 1) | (df['HS_Individual'] == 1) | (df['HS_Group'] == 1)]

# concatenate the cleaned_tweet values into a single string
text = ' '.join(map(str, selected_rows['cleaned_tweet'].tolist()))

# create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Set2').generate(text)

# display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Hate-Speech for tweeter target')
plt.show()
```
